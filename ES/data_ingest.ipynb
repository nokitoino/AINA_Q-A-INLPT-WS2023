```python
input_qns = 'What is the projected number of diabetic patients by 2034 according to the International Diabetes Federation?'
vector_of_input_question = model.encode(input_qns)


# Step 1: Full-text search component
# Adjusted Full-text search component
text_query = {
    'multi_match': {
        'query': input_qns,
        'fields': ['title.full_text', 'abstract.full_text'],
        'type': 'cross_fields',
        'operator': 'or'  # Adjusted to 'or' for broader matching
    }
}

# Execute the full-text search
text_search_res = es.search(index='med_docs3', body={'query': text_query}, size=5)
print("Full-text Search Results:\n", text_search_res['hits']['hits'])

# Step 2: Vector search component (KNN)
vector_query = {
    'script_score': {
        'query': {'match_all': {}},
        'script': {
            'source': "cosineSimilarity(params.query_vector, 'abstract_encoded') + 1.0",
            'params': {'query_vector': vector_of_input_question}
        }
    }
}

# Execute the vector search
vector_search_res = es.search(index='med_docs3', body={'query': vector_query}, size=5)
print("\nVector Search Results:\n", vector_search_res['hits']['hits'])
```

```python
# Step 3: Combine both in a bool query
if text_search_res['hits']['hits'] and vector_search_res['hits']['hits']:
    combined_query = {
        'bool': {
            'must': text_query,
            'should': [
                {
                    'script_score': {
                        'query': {'match_all': {}},
                        'script': {
                            'source': "cosineSimilarity(params.query_vector, 'abstract_encoded') + 1.0",
                            'params': {'query_vector': vector_of_input_question}
                        }
                    }
                }
            ],
            'minimum_should_match': 1
        }
    }
    combined_res = es.search(index='med_docs3', body={'query': combined_query}, size=5)
    print("\nCombined Search Results:\n", combined_res['hits']['hits'])
```
